# Exploración Archivos Google maps
Se busca entender la naturaleza, formato y estructura que contienen los archivos de metadatada de google maps para luego realizar un proceso de ETL.

```python
#Importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
#leemos el archivo en formato json
dir = ('data\metadata-sitios/1.json')
uno = pd.read_json(dir,lines=True)
```

```python
#Se observa el dataframe
uno.sample(5)
```

```python
#Obtenemos informacion del dataframe
uno.info()
```

```python
#Se decide observar la columna "Category" con detenimiento.
```

```python
#Se crea la variable que nos muestre las categorias
category = uno['category'].dropna().explode()
#Se guardan las categorias unicas en la variable.
category = pd.Series(category.unique())

category
```

```python
#Se convierten el tipo de datos de la columna category en str
uno['category'] = uno['category'].str.replace(r' | ','',regex =True)
```
