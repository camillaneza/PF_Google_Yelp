```python
import pandas as pd
import os

# Ruta al archivo JSON grande
input_json_file = "/content/drive/MyDrive/yelp/review.json"

# Directorio donde se guardarán las partes divididas
output_directory = "/content/drive/MyDrive/Review-dividido"

# Crea el directorio de salida si no existe
os.makedirs(output_directory, exist_ok=True)

# Tamaño máximo de cada parte
max_chunk_size = 200000  # por ejemplo, dividir en partes de 500,000 filas

# Lee el archivo JSON grande en un DataFrame de pandas
df = pd.read_json(input_json_file, lines=True, chunksize=max_chunk_size)

part_number = 0  # Inicializa el número de parte

for chunk in df:
    part_number += 1  # Incrementa el número de parte en cada iteración
    output_file = os.path.join(output_directory, f"part_{part_number}.json")
    chunk.to_json(output_file, orient="records", lines=True)
```

```python
import pandas as pd
from IPython.display import display

# Lista de rutas a los archivos JSON
file_paths = [
    "/content/drive/MyDrive/Review-dividido/part_1.json",
    "/content/drive/MyDrive/Review-dividido/part_2.json",
    "/content/drive/MyDrive/Review-dividido/part_3.json"
]

# Leer cada archivo JSON en un DataFrame de pandas y almacenarlo en una lista
dataframes = [pd.read_json(file_path, lines=True) for file_path in file_paths]

# Concatenar los DataFrames en uno solo
reviewYelp1_csv = pd.concat(dataframes, ignore_index=True)

# Guardar el DataFrame concatenado en un archivo CSV
output_csv = "/content/drive/MyDrive/yelpreviewYelp1.csv"
reviewYelp1_csv.to_csv(output_csv, index=False)

# Crea el directorio de salida si no existe
os.makedirs(output_directory, exist_ok=True)

display(reviewYelp1_csv)
```

```python
import pandas as pd
from IPython.display import display

# Ruta del archivo CSV original
archivo_csv = "/content/drive/MyDrive/yelpreviewYelp1.csv"

# Cargar el archivo CSV en un DataFrame de pandas
df = pd.read_csv(archivo_csv)

# Modificar los tipos de datos de las columnas
df['text'] = df['text'].astype(str)
df['date'] = pd.to_datetime(df['date'])

# Filtrar por fechas a partir de 2017
filtro_fecha = df['date'].dt.year >= 2017
df_filtrado = df[filtro_fecha]

# Ruta para guardar el nuevo archivo CSV
archivo_csv_filtrado = "ReviewYelp1ro.csv"

# Guardar el DataFrame filtrado en un nuevo archivo CSV
df_filtrado.to_csv(archivo_csv_filtrado, index=False)

# Mostrar el DataFrame filtrado usando display
display(df_filtrado)
```
