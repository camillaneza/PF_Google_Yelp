## Se concatenan todos los archivos ##


```python
import pandas as pd

# Lista de rutas a los archivos CSV a concatenar
rutas_csv = [
    "/content/drive/MyDrive/Review1.csv",
    "/content/drive/MyDrive/Review2.csv",
    "/content/drive/MyDrive/Review3.csv",
    "/content/drive/MyDrive/Review4.csv",
    "/content/drive/MyDrive/Review5.csv",
    "/content/drive/MyDrive/Review6.csv",
    "/content/drive/MyDrive/Review7.csv"
]

# Lista para almacenar DataFrames de cada archivo CSV
dataframes = []

# Cargar y concatenar los archivos CSV
for ruta_csv in rutas_csv:
    df = pd.read_csv(ruta_csv)
    dataframes.append(df)

# Concatenar los DataFrames en uno solo
df_concatenado = pd.concat(dataframes, ignore_index=True)

# Ruta al archivo CSV resultante
ruta_csv_resultante = "/content/drive/MyDrive/ReviewYelp.csv"

# Guardar el DataFrame concatenado en un archivo CSV llamado "ReviewYelp.csv"
df_concatenado.to_csv(ruta_csv_resultante, index=False)

# Mostrar el DataFrame concatenado utilizando display
display(df_concatenado)
```

```python
import pandas as pd
import re

# Ruta del archivo CSV
ruta_archivo = "/content/drive/MyDrive/ReviewYelp.csv"

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(ruta_archivo)

# Aplicar re.sub solo si el valor es una cadena
df['text'] = df['text'].apply(lambda x: re.sub(r'\([^)]*\)|[\u4e00-\u9fff]', '', x) if isinstance(x, str) else x)

df['text'] = df['text'].apply(lambda x: x.split('\n\n\n')[0].strip() if isinstance(x, str) else x)
```

```python
import pandas as pd

# Ruta al archivo pickle
ruta_pkl = "/content/drive/MyDrive/yelp/business.pkl"

# Cargar el archivo pickle en un DataFrame
df = pd.read_pickle(ruta_pkl)

# Eliminar columnas duplicadas manteniendo solo la primera apariciÃ³n de cada una
df = df.loc[:, ~df.columns.duplicated(keep='first')]

# Guardar el DataFrame en un archivo pickle llamado 'business.pkl' (correcciÃ³n en la ruta)
ruta_pkl_corregida = "/content/drive/MyDrive/business.pkl"
df.to_pickle(ruta_pkl_corregida, protocol=4)  # Usamos protocol=4 para garantizar la compatibilidad entre versiones

# Mostrar las columnas restantes
print(df.columns)
```

## Realizamos la sustituciÃ³n de los cÃ³digos de estado al nombre completo.##

## El DataFrame de 2017 a 2022 lo dividimos en dos tramos para el anÃ¡lisis posterior y carga.

```python
import pandas as pd

# Ruta al archivo CSV original
ruta_csv = "/content/drive/MyDrive/YelpRestaurantes4.csv"

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(ruta_csv)

# Convertir la columna "Anio" a tipo datetime
df['anio'] = pd.to_datetime(df['anio'], format='%Y')

# Filtrar datos anteriores a 2021
df_2017_2020 = df[df['anio'] < '2021']

# Filtrar datos desde 2021 en adelante
df_desde_2021 = df[df['anio'] >= '2021']

# Ruta a los archivos CSV resultantes
ruta_csv_2017_2020 = "/content/drive/MyDrive/YelpRestaurantes2017-2020.csv"
ruta_csv_desde_2021 = "/content/drive/MyDrive/YelpRestaurantesdesde2021.csv"

# Guardar los DataFrames filtrados en los archivos CSV correspondientes
df_2017_2020.to_csv(ruta_csv_2017_2020, index=False)
df_desde_2021.to_csv(ruta_csv_desde_2021, index=False)
```
