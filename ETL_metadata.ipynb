```python
import pandas as pd
import numpy as np
import json
```

```python

# Crear un diccionario para almacenar los DataFrames
dataframes = {}
#Se crea el filtro
filtros = ['Restaurant']
# Importar los DataFrames desde los archivos JSON
for i in range(1, 12):
    nombre_archivo = f'data/metadata-sitios/{i}.json'
    nombre_dataframe = f'uno_{i}'  # Puedes ajustar el nombre del DataFrame según tus necesidades
    dataframes[nombre_dataframe] = pd.read_json(nombre_archivo, lines=True)

# Crear una lista para almacenar los DataFrames filtrados
dataframes_filtrados = []

# Iterar sobre los DataFrames
for nombre_dataframe, df_actual in dataframes.items():
    # Convertir la columna 'category' a cadena
    df_actual['category'] = df_actual['category'].astype(str)

    # Filtrar el DataFrame para las filas donde 'categories' no es NaN
    df_filtrado = df_actual.dropna(subset=['category'])

    # Aplicar el filtro
    df_filtrado = df_filtrado[df_filtrado['category'].str.contains('|'.join(filtros), regex=True)]

    # Agregar el DataFrame filtrado a la lista
    dataframes_filtrados.append(df_filtrado)

# Concatenar los DataFrames filtrados en uno solo
df_resultado_metadata = pd.concat(dataframes_filtrados, ignore_index=True)
```
